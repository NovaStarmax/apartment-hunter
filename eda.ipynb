Transformer floor :

Bajo (Rez-de-chaussée) -> 0

Semi-sótano (Demi sous-sol) -> -1

Sótano (Sous-sol) -> -2

Entreplanta (Entresol) -> 0.5

- Les Surfaces

**sq_mt_built**
GARDER
Pourquoi : C'est la variable la plus corrélée au prix. Indispensable.

**sq_mt_useful**
SUPPRIMER
Pourquoi : Extrêmement corrélée à sq_mt_built (colinéarité). Avoir les deux n'apporte rien au modèle et peut le perturber. On garde la "built" car elle est plus souvent renseignée.

**sq_mt_allotment**
SUPPRIMER
Pourquoi : Presque vide pour les appartements (majorité du dataset). Utile seulement pour les maisons isolées, trop de valeurs manquantes (NaN) ici.

- Le Prix
**buy_price**
GARDER Cible 

**buy_price_by_area** 
SUPPRIMER
Pourquoi : C'est une fuite de données (Data Leakage). Cette colonne contient la réponse (buy_price / sq_mt). Si tu la gardes, ton modèle trichera et aura un score de 100%, mais ne servira à rien.

**rent_price, rent_price_by_area, is_rent_price_known**
SUPPRIMER
Pourquoi : Concerne la location, hors sujet pour la vente.

- Les Commodités
**Parking (has_parking, has_private..., has_public..., parking_price)**
TRANSFORMER
Détail : Créer une seule colonne has_parking (True/False). Si n'importe laquelle des colonnes parking est vraie, alors c'est Vrai. On supprime les détails (parking_price est trop souvent vide).

**Chauffage / Clim (has_central_heating, has_individual..., has_ac)**
GARDER has_ac, FUSIONNER les chauffages.
Pourquoi : has_ac (Clim) est un luxe qui monte le prix. Pour le chauffage, peu importe s'il est central ou individuel pour une première analyse, l'important est d'en avoir un. On crée une colonne has_heating.

- État et Type du bien
**is_renewal_needed**
GARDER
Pourquoi : Impact négatif fort sur le prix.

**is_new_development**
GARDER
Pourquoi : Impact positif fort (immobilier neuf plus cher).

**built_year**
GARDER
Pourquoi : L'ancienneté joue sur le prix.
Transformation : Remplacer les NaN par la médiane (comme vu avant).

**energy_certificate**
TRANSFORMER
Pourquoi : Actuellement c'est du texte (A, B, C...). Un modèle préfère des chiffres ordonnés.
Transformation : Convertir en échelle numérique : A=1 (Top), G=7 (Pire), ou l'inverse. Les "non indiqué" seront la valeur médiane.

**raw_address, street_name, etc.**
SUPPRIMER
Pourquoi : Texte inutile.

**latitude, longitude**
SUPPRIMER
Pourquoi : Dans ce dataset précis, ces colonnes sont souvent vides ou bruitées car les sites d'annonces cachent l'adresse exacte. On se fie à tes colonnes neighborhood et district créées précédemment qui sont plus fiables.

- 

**neigboorh id**
Neighborhood 135: San Cristóbal (1308.89 €/m2) - Dist*rict 21: Villaverde
ex : San Cristobal
    


```python
# Filtre : On ne garde que les lignes où le prix est strictement supérieur à 10 000.
df = df.dropna(subset=['buy_price'])
df = df[df['buy_price'] > 10000]

# Traitement de la colonne 'neighborhood_id' pour extraire quartier, district et prix moyen au m2
pattern = r"Neighborhood \d+: (?P<neighborhood>.*?) \((?P<neigh_price_m2>\d+\.?\d*) €/m2\) - District \d+: (?P<district>.*)"
extracted = df['neighborhood_id'].str.extract(pattern)
df['neighborhood'] = extracted['neighborhood'].str.strip()
df['district'] = extracted['district'].str.strip()
df['neigh_price_m2'] = pd.to_numeric(extracted['neigh_price_m2'])

# Mettre en ordinal la colonne Energy Certificate
# Pour les logements de moins de 50 m2 dont exempté de Classe energetique (inmueble exento) donc quoi mettre   
energy_map = {'A': 7, 'B': 6, 'C': 5, 'D': 4, 'E': 3, 'F': 2, 'G': 1}
df['energy_certificate'] = df['energy_certificate'].map(energy_map)
df['energy_certificate'] = df['energy_certificate'].fillna(df['energy_certificate'].median())
    
# Nettoyage de la colonne 'floor' retravailler
def clean_floor(val):
    if pd.isna(val): return np.nan
    val = str(val).lower()
    if 'bajo' in val: return 0
    if 'sótano' in val: return -1
    if 'entreplanta' in val: return 0.5
    try: return float(val)
    except: return np.nan

df['floor'] = df['floor'].apply(clean_floor)
df['floor'] = df['floor'].fillna(df['floor'].mode()[0])

# En cas de surface Null mettre la median
df['sq_mt_built'] = df['sq_mt_built'].fillna(df['sq_mt_built'].median())

# Remplacer les NaN par la médiane et filtrer pour ne garder que les années inférieures ou égales à l'année actuelle
df['built_year'] = df['built_year'].fillna(df['built_year'].median())
df = df[df['built_year'] <= datetime.date.today().year]

# Prendre uniquement le type de if 'house_type_id' in df.columns:
if 'house_type_id' in df.columns:
    df['house_type'] = df['house_type_id'].str.split(':').str[0].str.replace('HouseType ', '')
    
# final df 
final_cols = [
    'buy_price',            # Cible
    'sq_mt_built',          # Surface
    'n_rooms', 'n_bathrooms', # Pièces (Penser à imputer NaNs si besoin)
    'floor',                # Étage nettoyé
    'house_type',           # Type
    'built_year',           # Ancienneté
    'neighborhood', 'district', 'neigh_price_m2', # Localisation
    'energy_certificate' # Classe énergétique

]

df_final = df[final_cols].copy()

# Remplacement des derniers NaNs catégoriels
df_final['house_type'] = df_final['house_type'].fillna(df_final['house_type'].mode()[0])

# En cas de Pièce Null mettre la median
df_final['n_rooms'] = df_final['n_rooms'].fillna(df_final['n_rooms'].median())
df_final['n_bathrooms'] = df_final['n_bathrooms'].fillna(df_final['n_bathrooms'].median())

# Sauvegarde
df_final.to_csv('houses_Madrid_MASTER.csv', index=False)

print("Dataset MASTER créé : 'houses_Madrid_MASTER.csv'")
print(df_final.info())
```
